####  2) Faça um boxplot do peso do bebê

Resolvendo tarefa 2 - Exercicio 2

```python
df['IDADEMAE'].plot.hist(bins=20)
plt.title('Histograma de IDADEMAE')
plt.xlabel('Idade da mãe')
plt.show()

df['PESO'].plot.hist(bins=20)
plt.title('Histograma de PESO')
plt.xlabel('Peso do bebê')
plt.show()

df['APGAR1'].plot.hist(bins=20)
plt.title('Histograma de APGAR1')
plt.xlabel('APGAR1')
plt.show()

df['APGAR5'].plot.hist(bins=20)
plt.title('Histograma de APGAR5')
plt.xlabel('APGAR5')
plt.show()
```

####  3) Faça um histograma da idade da mãe

REsolvendo tarefa 2 - Exercicio 3

```python
sns.displot(data=df, x='IDADEMAE', bins=20, kde=True)
plt.title('Distribuição de IDADEMAE')
plt.show()

sns.displot(data=df, x='PESO', bins=20, kde=True)
plt.title('Distribuição de PESO')
plt.show()

sns.displot(data=df, x='APGAR1', bins=20, kde=True)
plt.title('Distribuição de APGAR1')
plt.show()

sns.displot(data=df, x='APGAR5', bins=20, kde=True)
plt.title('Distribuição de APGAR5')
plt.show()
```
